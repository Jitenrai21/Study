```python
import pandas as pd
```

```python
car_dataset = pd.read_csv('car(in).csv')
```

```python
car_dataset.head(10)
```

```python
car_dataset.drop(columns=['Types', 'Waranty', 'Used For'], inplace=True)
car_dataset.head()
```

```python
# Clean Price column - remove Rupee symbol and handle multiple prices
import re

def clean_price(price_str):
    if pd.isna(price_str):
        return None
    
    price_str = str(price_str)
    
    # Find all numbers in the format of currency (with commas)
    pattern = r'[\d,]+'
    matches = re.findall(pattern, price_str)
    
    # Convert matched strings to integers (remove commas)
    numbers = []
    for match in matches:
        cleaned = match.replace(',', '').strip()
        if cleaned.isdigit() and len(cleaned) > 2:  # Filter out small numbers
            numbers.append(int(cleaned))
    
    # If multiple prices found, return average; otherwise return the single price
    if len(numbers) > 1:
        return sum(numbers) / len(numbers)
    elif len(numbers) == 1:
        return numbers[0]
    else:
        return None

# Apply the cleaning function
car_dataset['Price_Cleaned'] = car_dataset['Price'].apply(clean_price)

# Check the results
print("Original vs Cleaned Price (first 10):")
print(car_dataset[['Price', 'Price_Cleaned']].head(10))
```

```python
car_dataset['Price_Cleaned'].describe()
```

```python
# Step 1: Standardize the Price_Cleaned column using StandardScaler
from sklearn.preprocessing import StandardScaler
import numpy as np

# Create a StandardScaler instance
scaler = StandardScaler()

# Fit and transform only non-null values
# Standardization formula: z = (x - mean) / std
price_standardized = scaler.fit_transform(car_dataset[['Price_Cleaned']].dropna())

# Add standardized values to the dataset
car_dataset['Price_Standardized'] = np.nan
car_dataset.loc[car_dataset['Price_Cleaned'].notna(), 'Price_Standardized'] = price_standardized

print("="*80)
print("STANDARDIZATION RESULTS")
print("="*80)
print(f"\nOriginal Price_Cleaned statistics:")
print(f"  Mean: ₹{car_dataset['Price_Cleaned'].mean():,.2f}")
print(f"  Std Dev: ₹{car_dataset['Price_Cleaned'].std():,.2f}")

print(f"\nStandardized Price statistics:")
print(f"  Mean: {car_dataset['Price_Standardized'].mean():.6f} (should be ~0)")
print(f"  Std Dev: {car_dataset['Price_Standardized'].std():.6f} (should be ~1)")

print(f"\n{'='*80}")
print("Sample of Original vs Standardized Prices:")
print(f"{'='*80}")
print(car_dataset[['Price', 'Price_Cleaned', 'Price_Standardized']].head(15))
```
