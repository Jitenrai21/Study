# Comprehensive EDA Summary: Nepal Used Car Market 2025 (847 Vehicles)

## 1. Dataset Overview
- **Total listings**: 847 fully cleaned vehicles  
- **Price range**: NPR 1.5 Lakh → NPR 2.4 Crore  
- **Median price**: **NPR 17.7 Lakh** (true "typical" used car price in Nepal)  
- **Average price**: NPR 23.42 Lakh (pulled up by Toyota SUVs & EVs)  
- **Year range**: 1975–2023 | **Median year**: 2014 | **Average year**: 2013.3  
- **Average odometer**: 51,112 km  
- **Average mileage**: 18.91 km/l  
- **Average engine**: 1,524 CC  
- **100% complete data** — no missing values after cleaning

### Data Quality: Missing Values Analysis
![Missing Data Analysis](../DataVisualisationAssets/Missing_Data_Analysis_Original_Dataset.png)

The original dataset had significant missing values that were systematically cleaned through description mining and KNN imputation, resulting in a 100% complete dataset ready for analysis.

## 2. Market Composition

### Brand Distribution
![Brand Distribution](../DataVisualisationAssets/Brand_Distribution.png)

The market is dominated by **Hyundai** (largest volume), followed by **Maruti Suzuki** and **Toyota**. Despite not having the highest count, Toyota commands the highest average prices due to its premium SUV lineup.

### Fuel Type Distribution  
![Fuel Type Distribution](../DataVisualisationAssets/Fuel_Type_Distribution.png)

**Petrol** vehicles dominate the market (71%), followed by **Diesel** (26%). Electric and Hybrid vehicles remain niche but command premium pricing.

### Transmission Distribution
![Transmission Distribution](../DataVisualisationAssets/Transmission_Distribution.png)

**Manual 2WD** transmissions dominate (61%), reflecting Nepal's preference for affordable, fuel-efficient vehicles. Automatic and 4WD variants are premium segments.

**Answer**: Yes — fuel type is a **huge** price driver.

### Are electric vehicles priced significantly different from petrol?
**Yes — dramatically.**  
- Average EV price: **NPR 41.3 Lakh**  
- Average petrol price: **NPR 20.6 Lakh**  
→ **100%+ premium** for electric despite lower running costs and smaller engines  
→ Mostly Tata Nexon EV & MG ZS EV.

### What's the relationship between engine size and price?
**Extremely strong** — the clearest pattern in the entire dataset.

| Engine Category       | Count | Avg Price     | Price Increase |
|-----------------------|-------|---------------|----------------|
| Small (≤1200 CC)      | 406   | 17.3 Lakh     | Baseline       |
| Medium (1201–2000 CC) | 300   | 24.5 Lakh     | **+41%**       |
| Large (>2000 CC)      | 141   | 38.8 Lakh     | **+124%**      |

**Rule of thumb**: Every additional 800–1000 CC adds ~NPR 8–12 Lakh in resale value  
→ Engine size beats year, mileage, and even brand in many cases.

### Do manual transmissions really have better mileage?
**Yes — but only slightly.**

| Transmission Type | Avg Mileage (km/l) | Count | Avg Price Premium |
|-------------------|--------------------|-------|-------------------|
| Manual2WD         | **19.4**           | 515   | Baseline          |
| Auto2WD           | 17.8               | 131   | +15–20%           |
| Manual4WD         | 15.1               | 108   | +30–50%           |
| Auto4WD           | 14.3               | 48    | +60–100%          |

Manuals win on fuel economy, but buyers pay big premiums for automatic convenience and 4WD capability.

### Which transmission type commands premium prices?
| Type      | Avg Price Premium vs Manual2WD | Reason |
|-----------|-------------------------------|--------|
| Auto4WD   | **+80–100%**                  | Luxury SUVs (Fortuner, Pajero) |
| Manual4WD | +40–60%                       | Off-road credibility |
| Auto2WD   | +15–25%                       | Comfort in city cars |

**Auto4WD is the ultimate premium combo** — found mostly on Toyota SUVs.

### Is there a sweet spot for year / mileage / price?
**YES — very clearly defined.**

**The "Smart Buyer Sweet Spot" (276 vehicles — 32% of market)**  
- Engine: 1,000 – 1,800 CC  
- Odometer: ≤100,000 km  
- Price: ≤ NPR 17.7 Lakh (median)  
- **Average price in sweet spot: NPR 12.04 Lakh**  
- Dominant combo: **Petrol + Hyundai/Maruti Suzuki + Manual**

These are the cars that give you the **most engine per rupee** with reasonable usage — the true "bang for buck" zone.

## Final Key Market Insights — Nepal Used Car Market 2025

| Insight | Detail |
|-------|--------|
| 1. Engine size > mileage > brand in price impact | Bigger engines win — Nepali buyers prioritize power |
| 2. **Toyota has god-tier resale** | Even old models beat newer Korean cars |
| 3. **Hybrids & EVs are luxury items** | 2–3× price despite low volume |
| 4. Diesel holds value surprisingly well | Despite higher mileage |
| 5. **Manuals are more efficient**, autos are more expensive | Convenience tax is real |
| 6. **Sweet spot = 1.0–1.8L petrol under 1 lakh km** | ~12 Lakh average — best value zone |
| 7. 4WD + Automatic = ultimate premium combo | Found on high-end Toyotas |
| 8. High-mileage diesels & Tata cars often underpriced | Hidden bargain opportunities |
| 9. **Odometer matters less than expected** | Brand and engine size dominate pricing |
| 10. **Post-2020 price spike** | EVs and supply chain impacts visible |

**Bottom line**:  
The Nepal used car market strongly rewards **big engines, Toyota badges, diesel/hybrid powertrains, and automatic/4WD convenience** — while being surprisingly forgiving on odometer readings.  
The smartest buys live in the **1.0–1.8L petrol sweet spot under 17.7 Lakh** — but if you want status and long-term value, nothing beats a Toyota with a big engine.
